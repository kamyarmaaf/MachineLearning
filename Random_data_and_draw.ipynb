```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(90)  # برای تولید داده‌های قابل تکرار  
X = np.linspace(0, 8, 30)  # 50 مقدار بین 0 تا 120  
y = X**2 + 1 + np.random.randn(30) * 3  # خط y = 2x + 1با نویز تصادفی

%matplotlib inline

plt.scatter(X, y, color="orange", label="Data points")  # رسم نقاط تصادفی  
plt.plot(X, X**2+1, color="black", label="True line")  # خط واقعی بدون نویز  
plt.xlabel("X values")
plt.ylabel("y values")
plt.legend()
plt.title("Random Data with Noise")
plt.show()
```
