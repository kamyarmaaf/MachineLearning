```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0.1, 5, 100)
y_exp = np.exp(X)
y_log = np.log(X)
y_exp_deriv = np.exp(X)
y_log_deriv = 1/X

plt.figure(figsize=(8, 5))
plt.plot(X, y_exp, label="y = e^X", color="red", linestyle="solid")
plt.plot(X, y_exp_deriv, label="y = d(e^X)/dx", color="yellow", linestyle="dashed")
plt.plot(X, y_log, label="y = ln(X)", color="blue", linestyle="solid")
plt.plot(X, y_log_deriv, label="y = d(log(X)/dx)", color="pink", linestyle="solid")
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("exponential & log functions")
plt.legend()
plt.grid()
plt.show()
```
