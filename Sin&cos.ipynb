```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
X=np.linspace(0,2*np.pi,200)
Y1=np.sin(X)
Y2=np.cos(X)
dY1=np.cos(X)
dY2=-np.sin(X)
plt.plot(X,Y1,color="blue",linestyle="dashed",label="sin(x)")
plt.plot(X,Y2,color="red",linestyle="solid",label="cos(x)")
plt.plot(X,dY1,color="green",linestyle="dashed",label="d(sin(X))/d(X)=cos(X)")
plt.plot(X,dY2,color="purple",linestyle="solid",label="d(cos(X))/d(X)=-sin(X)")


plt.xlabel("x values")
plt.ylabel("y values")
plt.title("diagrom of sinx & cosx & derivatives")
plt.legend()
plt.grid()
plt.show()
```
