```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.normal(loc=5, scale=1.5, size=1000)
data2 = np.random.normal(loc=10, scale=2.5, size=1000)
plt.hist(data1, bins=20, color='red', alpha=0.6, edgecolor='pink', label='mean=5, std=1.5')
plt.hist(data2, bins=20, color='blue', alpha=0.6, edgecolor='white', label='mean=10, std=2.5')
plt.xlabel('meqdar')
plt.ylabel('tedad')
plt.title('conversion')
plt.legend()
plt.show()
```
